```python

#!/usr/bin/python
# -*- coding:utf-8 -*-
import pandas as pd
import numpy as np
# import networkx as nx
import matplotlib.pyplot as plt
from datetime import datetime
from collections import Counter
from sklearn.linear_model import LinearRegression
from scipy import stats
from sklearn.linear_model import RANSACRegressor
import scipy.stats
from matplotlib import style
from datetime import timedelta
from scipy.stats import chi2_contingency
# import seaborn as sns
from matplotlib import style

Net_df = pd.read_csv('Data/Net_df_1223_total_region.csv')
style.use('ggplot')

Net_df['inf_time'] = pd.to_datetime(Net_df['inf_time'], format="%Y-%m-%d", errors='ignore')
Net_df['inf_time'] = Net_df['inf_time'].dt.date
Net_df['Primary_id'] = Net_df['Primary_id'].astype(int)
```

```python
plt.rcParams["font.family"] = "Palatino Linotype"

target_date = ['2020-01-19', '2020-04-30', '2020-07-15', '2020-10-13', '2021-02-26', '2021-07-12', '2021-09-16']
# target_date = ['2020-01-19', '2021-09-16']
# target_date = ['2020-01-19', '2020-04-30']
cross_list = []
save_cross = pd.DataFrame({})
date_diff_U = []
date_diff_K = []
# frac = [0.2,0.3,0.2,0.05,0.05,0.05]
for i in range(len(target_date)):
    target_date[i] = datetime.strptime(target_date[i], "%Y-%m-%d").date()

for i in range(len(target_date) - 1):
    Net_df_1 = Net_df[(Net_df['inf_time'] >= target_date[i]) & (Net_df['inf_time'] < target_date[i + 1])]

    ## Calculate Delay
    date_diff_K.append([])
    date_diff_U.append([])    

    for i2 in Net_df_1.index:
        if len(Net_df_1.loc[i2, 'Onset_date']) != 1:
            Net_df_1.loc[i2, 'Onset_date'] = datetime.strptime(Net_df_1.loc[i2, 'Onset_date'][0:10], "%Y-%m-%d").date()

    Net_df_Onset_0 = Net_df_1.loc[:][Net_df_1.loc[:, 'Onset_date'] == '0'].index
    Net_df_3 = Net_df_1.drop(Net_df_Onset_0)

    date_diff_R1 = []
    date_degree_R1 = []
    for i2 in Net_df_3.index:
        
        date_to_compare2 = Net_df_3.loc[i2, 'Onset_date']
        date_to_compare1 = Net_df_3.loc[i2, 'inf_time']
        
        if Net_df_3.loc[i2,'Primary_id'] == 0:
            date_diff_U[i].append((date_to_compare1 - date_to_compare2).days)
            
        elif Net_df_3.loc[i2,'Primary_id'] != 0:
            date_diff_K[i].append((date_to_compare1 - date_to_compare2).days)
        
    print(str(i+1) + ' th Period')
    print('Unknown_Delay_mean : {0:.3}'.format(np.mean(date_diff_U[i])))
    print('known_Delay_mean : {0:.3}'.format(np.mean(date_diff_K[i])))
            
            
    ## Chi Sqare
    Net_df_2 = Net_df_1.copy()
    
    Primary_OX = []
    Age_OX = []
    Onset_OX = []
    Gender_OX = []
    for k in Net_df_2.index:
        if Net_df_2.loc[k,'Primary_id'] != 0:
            Primary_OX.append(1)
        elif Net_df_2.loc[k,'Primary_id'] == 0:
            Primary_OX.append(0)
            
            
        if Net_df_2.loc[k,'age'] < 20:
            Age_OX.append(10)
        elif 20 <= Net_df_2.loc[k,'age'] < 30:
            Age_OX.append(2030)
        elif 30 <= Net_df_2.loc[k,'age'] < 40:
            Age_OX.append(3040)
        elif 40 <= Net_df_2.loc[k,'age'] < 60:
            Age_OX.append(4060)
        elif 60 <= Net_df_2.loc[k,'age'] < 80:
            Age_OX.append(6080)
        elif 80 <= Net_df_2.loc[k,'age']:
            Age_OX.append(9000)
            
            
        if Net_df_2.loc[k,'Onset_date'] != '0':
            Onset_OX.append('Onset')
        elif Net_df_2.loc[k,'Onset_date'] == '0':
            Onset_OX.append('Asym')
            
            
        if Net_df_2.loc[k,'gender'] == '남':
            Gender_OX.append('Man')
        elif Net_df_2.loc[k,'gender'] == '여':
            Gender_OX.append('Women')

            
    Net_df_2['Primary_OX'] = Primary_OX
    Net_df_2['Age_OX'] = Age_OX        
    Net_df_2['Onset_OX'] = Onset_OX
    Net_df_2['Gender_OX'] = Gender_OX
    gender_cross = pd.crosstab(Net_df_2['Gender_OX'], Net_df_2['Primary_OX'])#, normalize='columns')
    age_cross = pd.crosstab(Net_df_2['Age_OX'], Net_df_2['Primary_OX'])#, normalize='columns')
    onset_cross = pd.crosstab(Net_df_2['Onset_OX'], Net_df_2['Primary_OX'])#, normalize='columns')
    
    gender_cross_norm = pd.crosstab(Net_df_2['Gender_OX'], Net_df_2['Primary_OX'], normalize='columns')
    age_cross_norm = pd.crosstab(Net_df_2['Age_OX'], Net_df_2['Primary_OX'], normalize='columns')
    onset_cross_norm = pd.crosstab(Net_df_2['Onset_OX'], Net_df_2['Primary_OX'], normalize='columns')
    print(age_cross.sum())
    # cross_list.append([gender_cross, age_cross, onset_cross])
    
    save_cross_tmp_1 = pd.concat([age_cross, age_cross_norm], axis=1)
    save_cross_tmp_2 = pd.concat([gender_cross, gender_cross_norm], axis=1)
    save_cross_tmp_3 = pd.concat([onset_cross,onset_cross_norm], axis=1)
    
    
    save_cross = pd.concat([save_cross, pd.concat([save_cross_tmp_1, save_cross_tmp_2, save_cross_tmp_3])], axis=1)

    heatmap_list = [gender_cross, age_cross, onset_cross]
    print_tmp = ['gender', 'age', 'onset']
    for i2 in range(len(heatmap_list)):
        # plt.figure(i2,figsize=(12,8))
        # sns.heatmap(heatmap_list[i2], annot=True, cmap="YlGnBu")
        # plt.show()
        c, p, dof, expected = chi2_contingency(heatmap_list[i2])
        print(print_tmp[i2] + ' P-value : {:.5f}'.format(p)) # p-value가 낮으면 통계적으로 유의함. 관련이 있음.
        # print('P-value :' + str(p)) # p-value가 낮으면 통계적으로 유의함. 관련이 있음.
    print('Length : ' + str(len(Net_df_2)))
    print(' ')
    
    df123 = pd.crosstab(Net_df_2['Age_OX'], Net_df_2['Primary_OX'], normalize='columns')
    

    ## Real Age Distribution
    
    Age_dist = pd.read_csv('Data/Age_distribution_Montly.csv',encoding='cp949')

    Age_dist_ratio_df = pd.DataFrame({'<20' : [],
                                    '20-29' : [],
                                    '30-39' : [],
                                    '40-59' : [],
                                    '60-79' : [],
                                    '>79' : []})

    for i2 in range(len(Age_dist)):
        Age_dist_ratio = np.array(Age_dist.loc[i2][2:]) / Age_dist.loc[i2][1]
        

        Age_dist_ratio_df.loc[i2] = [sum(Age_dist_ratio[0:2]),
                                    Age_dist_ratio[2],
                                    Age_dist_ratio[3],
                                    sum(Age_dist_ratio[4:6]),
                                    sum(Age_dist_ratio[6:8]),
                                    sum(Age_dist_ratio[8:11])]
                                    
    Period_Age_Dist = [np.sum(Age_dist_ratio_df.loc[:4])/5,
                       np.sum(Age_dist_ratio_df.loc[4:7])/4,
                       np.sum(Age_dist_ratio_df.loc[7:10])/4,
                       np.sum(Age_dist_ratio_df.loc[10:14])/5,
                       np.sum(Age_dist_ratio_df.loc[14:19])/6,
                       np.sum(Age_dist_ratio_df.loc[19:])/2]

    fig = plt.figure(figsize=(9,9))
    ax = fig.subplots()
    # print(np.array(df123[0])-np.array(Period_Age_Dist[i]))
    # print(np.array(df123[1])-np.array(Period_Age_Dist[i]))
    # plt.plot(np.array(df123[0])-np.array(Period_Age_Dist[i]), 'o', linewidth=10, color= '#F8766D')
    # plt.plot(np.array(df123[1])-np.array(Period_Age_Dist[i]), 'o', linewidth=5, color= '#619CFF')
    
    plt.bar([0.9,1.9,2.9,3.9,4.9,5.9],np.array(df123[0])-np.array(Period_Age_Dist[i]),width=0.2)
    plt.bar([1.1,2.1,3.1,4.1,5.1,6.1],np.array(df123[1])-np.array(Period_Age_Dist[i]),width=0.2)
    
    # plt.plot(Period_Age_Dist[i], linewidth=5)
    plt.xticks([1,2,3,4,5,6])
    plt.yticks(fontsize=20)
    ax.set_xticklabels(['<20', '20-29', '30-39', '40-59', '60-79', '>79'], fontsize=20)
    plt.legend(['Unlinked','Linked','Age_Dist'], fontsize=15)
    # plt.vlines([0,1,2,3,4,5], 0, np.array(df123[0])-np.array(Period_Age_Dist[i]), lw=10, color= '#F8766D')#, alpha = 0.7)
    # plt.vlines([0,1,2,3,4,5], 0, np.array(df123[1])-np.array(Period_Age_Dist[i]), lw=5, color= '#619CFF')#, alpha = 0.7)
    # plt.ylim([0, 0.5])
    plt.ylabel('Relative rate',fontsize=20)
    plt.xlabel('Age',fontsize=20)
    plt.ylim([-0.15, 0.15])
    plt.title('P' + str(i+1) , fontsize=20)
    plt.savefig(str(i+1) + ' Relative distribution of confirmed cases by age with respect to population distribution', bbox='tight')
    save_cross.to_csv('cross_table.csv',float_format = '%.2f')


```
